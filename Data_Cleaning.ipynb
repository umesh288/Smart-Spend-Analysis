```python
pr["Rent"]=pr["Rent"].round(2)
pr["Loan_Repayment"]=pr["Loan_Repayment"].round(2)
pr["Insurance"]=pr["Insurance"].round(2)
pr["Groceries"]=pr["Groceries"].round(2)
pr["Transport"]=pr["Transport"].round(2)
pr["Eating_Out"]=pr["Eating_Out"].round(2)
pr["Entertainment"]=pr["Entertainment"].round(2)
pr["Utilities"]=pr["Utilities"].round(2)
pr["Healthcare"]=pr["Healthcare"].round(2)
pr["Education"]=pr["Education"].round(2)
pr["Miscellaneous"]=pr["Miscellaneous"].round(2)
pr["Desired_Savings_Percentage"]=pr["Desired_Savings_Percentage"].round(2)
pr["Desired_Savings"]=pr["Desired_Savings"].round(2)
pr["Disposable_Income"]=pr["Disposable_Income"].round(2)
pr["Potential_Savings_Groceries"]=pr["Potential_Savings_Groceries"].round(2)
pr["Potential_Savings_Transport"]=pr["Potential_Savings_Transport"].round(2)
pr["Potential_Savings_Eating_Out"]=pr["Potential_Savings_Eating_Out"].round(2)
pr["Potential_Savings_Entertainment"]=pr["Potential_Savings_Entertainment"].round(2)
pr[['Potential_Savings_Utilities',"Potential_Savings_Healthcare", "Potential_Savings_Education", 'Potential_Savings_Miscellaneous']]=pr[['Potential_Savings_Utilities',"Potential_Savings_Healthcare", "Potential_Savings_Education", 'Potential_Savings_Miscellaneous']].round(2)
pr
```

```python
duplicates=pr.duplicated()
pr.drop_duplicates(keep='first', inplace=True)
duplicates
```

```python
pr.describe()
```

```python
pr = pr.rename(columns={'Loan_Repayment':'Loan Repayment', 'Eating_Out':'Eating Out', 'Desired_Savings_Percentage':'Desired Savings Percentage', 'Desired_Savings':'Desired Savings', 'Disposable_Income':'Disposable Income', 'Potential_Savings_Groceries':'Potential Savings Groceries', 'potential_savings_transport':'Potential Savings Transport', 'Potential_Savings_Eating_Out':'Potential Savings Eating Out', 'Potential_Savings_Entertainment':'Potential Savings Entertainment', 'Potential_Savings_Utilities':'Potential Savings Utilities', 'Potential_Savings_Healthcare':'Potential Savings Healthcare', 'Potential_Savings_Education':'Potential Savings Education', 'Potential_Savings_Miscellaneous':'Potential Savings Miscellaneous', 'Potential_Savings_Transport':'Potential Savings Transport'})
```

```python
pr
```

```python
unique_values= pr["Occupation"].unique()
unique_values
```

```python
rows_to_delete=pr[(pr[['Disposable Income', 'Age', 'Dependents', 'Rent', 'Loan Repayment','Insurance', 'Groceries', 'Transport', 'Eating Out', 'Entertainment', 'Utilities', 'Healthcare', 'Education', 'Miscellaneous', 'Desired Savings Percentage', 'Desired Savings', 'Potential Savings Groceries', 'Potential Savings Transport', 'Potential Savings Eating Out', 'Potential Savings Entertainment', 'Potential Savings Utilities', 'Potential Savings Healthcare', 'Potential Savings Education', 'Potential Savings Miscellaneous']]<0).any(axis=1)]
```

```python
pr.drop(rows_to_delete.index, inplace=True)
pr.drop(["Dependents", "City_Tier", 'Disposable Income'], inplace=True, axis=1)
```

```python
pr.sort_values(["Occupation", 'Age'], ascending=[True, True], inplace=True)
```

```python
pr
```

```python
pr["Savings"]= (pr["Income"] - pr["Rent"] - pr["Loan Repayment"] - pr["Insurance"] - pr["Groceries"] - pr["Transport"] - pr["Eating Out"] - pr["Entertainment"] - pr["Utilities"] - pr["Healthcare"] - pr["Education"] - pr["Miscellaneous"]).round(2)
```

```python
pr["Savings Percentage"] = ((pr["Savings"] / pr["Income"]) * 100).round(2)
pr
```

```python
pr['Income ID'] = range(1, len(pr) + 1)
pr.set_index('Income ID', inplace=True)
```
